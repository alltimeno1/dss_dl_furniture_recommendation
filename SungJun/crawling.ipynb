```python
url = 'https://ohou.se/store/category?category=0_2_0_6'
response = requests.get(url)
response
```

```python
items = response.json()['productions']
items
```

```python
# img 저장 

import os 

if not os.path.exists("datas"):
    os.makedirs("datas")
```

```python
def save_file(title,link):
    response = requests.get(link)
    with open ("datas/bed_{}.png".format(title),"wb") as f:
        f.write(response.content)
```

```python
for item in items:
    img_url = item['resized_image_url']
    item_id = item['id']
    save_file(item_id, img_url)
```
